```python
import pandas as pd
import numpy as np
```

```python
snv_annotations = pd.read_csv("../data/williams/signatures_dataset/DLPextra/SNV/snv_annotations.csv.gz")
```

```python
ov2295_annotations = snv_annotations[snv_annotations['patient'] == 'OV2295']
ov2295_annotations
```

```python
ov2295_annotations["mutation"] = ov2295_annotations["chr"] + ":" + ov2295_annotations["pos"].astype(str) + ":" + ov2295_annotations["ref"] + ":" + ov2295_annotations["alt"]
ov2295_annotations
```

```python
kmeans_labels = pd.read_csv("../data/laks/scope/kmeans_labels_filter_25_reclustered_all.csv", index_col=0)
kmeans_labels
```

```python
merged = pd.merge(kmeans_labels, ov2295_annotations, on="mutation", how="left")
merged
```

```python
mutations = [
    'MYH9',
    'TP53',
    'FOXP2',
    'SUGCT',
    'HTR1D',
    'INSL4',
    'CHD2',
    'ARID1A',
    'ZHX1',
]
```

```python
merged[merged['gene_name'].isin(mutations)][['clone', 'gene_name']]
```

```python
phertilizer_snv_clones = pd.read_csv("../phertilizer-outputs/laks/SNV_clusters.csv")
phertilizer_snv_clones['mutation'] = phertilizer_snv_clones["mutation"].str.split("_").str[1]
phertilizer_snv_clones
```
