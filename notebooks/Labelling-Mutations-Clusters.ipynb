## Mutated Genes from cBioPortal

```python
cbio = pd.read_csv("../data/laks/Mutated_Genes.txt", sep="\t")
cbio["frequency"] = cbio["Freq"].str.rstrip("%").astype(float)
cbio = cbio.sort_values("frequency", ascending=False)
cbio
```

```python
cbio_with_freq_gt = cbio[cbio['frequency'] >= 2.5]
cbio_with_freq_gt
```

```python
genes_with_freq_gt = cbio_with_freq_gt['Gene'].to_list()
genes_with_freq_gt
```

```python
merged3 = pd.merge(kmeans_labels, ov2295_annotations, on="mutation", how="left")
```

```python
merged_filt = merged3[merged3['gene_name'].isin(genes_with_freq_gt)][['clone', 'gene_name']]
merged_filt
```

```python
# merged_filt = merged_filt[merged_filt['clone'].isin([0, 10, 7])]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(merged_filt.value_counts(['gene_name', 'clone']).sort_index(level='gene_name'))
```
