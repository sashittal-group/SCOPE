# Simulation Data

```python
import pandas as pd
```

```python
STR = "n1000_m5_size100_cov0.1_p5_s0"

df_read_counts = pd.read_parquet(f"../data/simulation/ground_truth_with_loss/{STR}/sim_read_count.parquet")
df_read_counts
```

```python
df_total_long = df_read_counts.reset_index().melt(id_vars='index', var_name='mutation', value_name='total')
df_total_long = df_total_long.rename(columns={'index': 'cell_id'})
df_total_long
```

```python
df_variant_counts = pd.read_parquet(f"../data/simulation/ground_truth_with_loss/{STR}/sim_variant_count.parquet")
df_variant_counts
```

```python
df_var_long = df_variant_counts.reset_index().melt(id_vars='index', var_name='mutation', value_name='variant')
df_var_long = df_var_long.rename(columns={'index': 'cell_id'})
df_var_long
```

```python
df_long = pd.merge(df_total_long, df_var_long, on=['cell_id', 'mutation'])
df_long
```

```python
df_long = df_long[df_long['total'] != 0]
df_long['cell_id'] = df_long['cell_id'].str[1:].astype(int)
df_long['mutation'] = df_long['mutation'].str[1:].astype(int)

df_long
```

```python
df_mut_bin = pd.read_parquet(f"../data/simulation/ground_truth_with_loss/{STR}/sim_mutation_to_bin_mapping.parquet")
df_mut_bin.index.name = "mutation"

df_mut_bin
```

```python
import numpy as np
```

```python
df_mut_bin_unq = df_mut_bin.drop_duplicates(subset=["bin"], keep="first")
df_mut_bin_unq
```

```python
df_mut_bin_unq['bin_id'] = np.arange(len(df_mut_bin_unq))
df_mut_bin_unq
```

```python
df_mut_bin = pd.merge(df_mut_bin, df_mut_bin_unq, on='bin', how='left')
df_mut_bin.index.name = "mutation"
df_mut_bin
```

```python
df_long = pd.merge(df_long, df_mut_bin, on="mutation", how='left')
df_long
```

```python
import os

os.makedirs(f"../data/simulation/pharming_input/{STR}", exist_ok=True)
```

```python
pharming_read_counts = pd.DataFrame({
    'segment': df_long['bin_id'],
    'mutation': df_long['mutation'],
    'cell': df_long['cell_id'],
    'varReads': df_long['variant'],
    'totReads': df_long['total'],
})
pharming_read_counts.to_csv(f"../data/simulation/pharming_input/{STR}/read_counts.tsv", index=False, sep='\t')
pharming_read_counts
```

```python
df_copy_numbers_bins = pd.read_parquet(f"../data/simulation/ground_truth_with_loss/{STR}/sim_copy_numbers_of_segments.parquet")
df_copy_numbers_bins
```

```python
df_copy_numbers = pd.read_parquet(
    f"../data/simulation/ground_truth_with_loss/{STR}/sim_copy_numbers.parquet"
)

df_copy_numbers.index = df_copy_numbers.index.str[1:].astype(int)
df_copy_numbers.columns = df_copy_numbers.columns.str[1:].astype(int)

# long format
df_copy_numbers_long = (
    df_copy_numbers
    .reset_index()
    .melt(id_vars="index", var_name="mutation", value_name="copy")
    .rename(columns={"index": "cell"})
)

# attach bin id
df_copy_numbers_long_cn = df_copy_numbers_long.merge(
    df_mut_bin.reset_index()[["mutation", "bin_id"]],
    on="mutation",
    how="left"
)

# ---- KEY FIX: aggregate per cell / segment ----
df_segment_cn = (
    df_copy_numbers_long_cn
    .groupby(["cell", "bin_id"], as_index=False)["copy"]
    .mean()   # or max / first depending on your model
)

pharming_copy_numbers = pd.DataFrame({
    "segment": df_segment_cn["bin_id"],
    "cell": df_segment_cn["cell"],
    "copiesX": df_segment_cn["copy"],
    "copiesY": 0
})

# sanity check
assert not pharming_copy_numbers.duplicated(["cell", "segment"]).any()

pharming_copy_numbers.to_csv(
    f"../data/simulation/pharming_input/{STR}/copy_numbers.csv",
    index=False
)
```
